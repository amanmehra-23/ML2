Qn. Implement dimensionality reduction using **PCA**

1. Use the Iris Dataset present in the scikit-learn library
2. implement PCA to reduce dimension into 2 dimensions [PC1, PC2]
4. Display new reduced dimension values
5. Plot appropriate graph/ label w.r.t. principal component.

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler
```

```python
# Loading the dataset
iris = datasets.load_iris()

X = iris.data
y = iris.target
print("X:",X[0])
target_names = iris.target_names


# Scaling the data
scaler = MinMaxScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
# Define a function for plotting graphs
# Function for plotting graphs
def plot3clusters(X, y, title, vtitle):
    plt.figure()
    colors = ['navy', 'turquoise', 'darkorange']
    for color, i, target_name in zip(colors, [0, 1, 2], target_names):
        plt.scatter(X[y == i, 0], X[y == i, 1], color=color, label=target_name)
    plt.legend(loc='upper left')
    plt.title(title)
    plt.xlabel(vtitle + "1")
    plt.ylabel(vtitle + "2")
    plt.show()


# Implement and visualize PCA
pca = PCA(n_components=2)  # Reduce to 2 dimensions (PC1, PC2)
X_pca = pca.fit_transform(X_scaled)


# Display new reduced dimension values
print("PCA Reduced dimensions (PC1, PC2):")
print(X_pca[:5])  # Show first 5 rows of the reduced dataset

# Call the plot function to visualize the reduced dimensions
plot3clusters(X_pca, y, 'PCA 2D Visualization', 'Principal Component ')
```
